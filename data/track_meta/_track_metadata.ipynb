# preprocess_tracks
tracks.csv에서 각 정보를 세분화후 csv로 따로 저장

path: /home/namo/CNN/data/fma_metadata/track_meta

```python
csv_path = '/home/namo/CNN/data/fma_metadata/tracks.csv'
tracks = pd.read_csv(csv_path)
tracks = tracks.drop([1]).reset_index(drop=True)
tracks.loc[0, 'Unnamed: 0'] = 'id'
```

```python
tracks.head()
```

```python
tracks = tracks[tracks['set.1'].isin(['small', 'subset'])]
tracks.shape
```

```python
columns = tracks.columns

track_col = [col for col in columns if 'track' in col]
artist_col = [col for col in columns if 'artist' in col]
album_col = [col for col in columns if 'album' in col]
set_col = [col for col in columns if 'set' in col]
index_col = [col for col in columns if col == 'Unnamed: 0']

track = tracks[index_col + track_col]
artist = tracks[index_col + artist_col]
album = tracks[index_col + album_col]
set = tracks[index_col + set_col]
```

```python
track.columns = track.iloc[0]
track = track.drop([0]).reset_index(drop=True)
track = track.rename(columns={track.columns[0]: 'track_id'})
track.head()
```

```python
track.to_csv('/home/namo/CNN/data/fma_metadata/track_meta/track.csv', index=False)

artist.columns = artist.iloc[0]
artist = artist.drop([0]).reset_index(drop=True)
artist = artist.rename(columns={artist.columns[0]: 'track_id'})
artist.to_csv('/home/namo/CNN/data/fma_metadata/track_meta/artist.csv', index=False)

album.columns = album.iloc[0]
album = album.drop([0]).reset_index(drop=True)
album = album.rename(columns={album.columns[0]: 'track_id'})
album.to_csv('/home/namo/CNN/data/fma_metadata/track_meta/album.csv', index=False)

set.columns = set.iloc[0]
set = set.drop([0]).reset_index(drop=True)
set = set.rename(columns={set.columns[0]: 'track_id'})
set.to_csv('/home/namo/CNN/data/fma_metadata/track_meta/set.csv', index=False)
```
